```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def histogram_equalization(image, nbr_bins=256):
    # m bo hnh nh u vo l nh xm
    if image.mode != 'L':
        image = image.convert('L')
    
    # Chuyn i hnh nh thnh mng NumPy
    image_array = np.array(image)

    # Tnh ton histogram ca nh
    histogram, bins = np.histogram(image_array, bins=nbr_bins, range=(0, 256), density=True)

    # Tnh ton hm phn phi tch lu (CDF - Cumulative Distribution Function)
    cdf = histogram.cumsum()
    cdf = 255 * cdf / cdf[-1]

    # Ly gi tr mi cho tng pixel da trn CDF
    image_equalized = np.interp(image_array, bins[:-1], cdf)

    # Chuyn i mng kt qu thnh hnh nh
    equalized_image = Image.fromarray(image_equalized.astype('uint8'))

    return equalized_image
# c hnh nh t tp
my_path = 'C:/Users/DucLuong3312/Documents/TGMT/img/13.jpg'
image = Image.open(my_path)

# p dng cn bng lc  nh xm
equalized_image = histogram_equalization(image)

# To mt li 2x2  hin th hnh nh v biu 
plt.figure(figsize=(15, 10))

# Hin th hnh nh gc
plt.subplot(2, 2, 1)
plt.imshow(image.convert("L"), cmap='gray')
plt.title('Hnh nh gc')

# V biu  histogram ca hnh nh gc
plt.subplot(2, 2, 2)
plt.hist(np.array(image.convert("L")).flatten(), bins=128)
plt.title('Biu  Histogram - Hnh nh gc')

# Hin th hnh nh  cn bng
plt.subplot(2, 2, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title('Hnh nh  cn bng')

# V biu  histogram ca hnh nh  cn bng
plt.subplot(2, 2, 4)
plt.hist(np.array(equalized_image).flatten(), bins=128)
plt.title('Biu  Histogram - Hnh nh  cn bng')

plt.show()
```
